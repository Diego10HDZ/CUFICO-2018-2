```python
import ROOT
```

```python
from array import array
```

```python
MyTree = ROOT.TTree( 'tree1', 'Dos gaussianas y una plana' )
```

```python
MyTree = ROOT.TTree( 'tree1', 'Dos gaussianas y una plana' )
```

```python
Gauss1 = array( 'f', [ 0 ] )
Gauss2 = array( 'f', [ 0 ] )
Plana = array( 'f', [ 0 ] )
```

```python
MyTree.Branch( 'numero 1', Gauss1, 'Gauss1/F' )
MyTree.Branch( 'numero 2', Gauss2, 'Gauss2/F' )
MyTree.Branch( 'numero 3', Plana, 'Plana/F' )
```

```python
from random import randint
import numpy as np
```

```python
for i in range(1000):
  Gauss1[0] = np.random.normal(4,2)
  Gauss2[0] = np.random.normal(7,2)
  Plana[0] = np.random.uniform(0, 10) 
  
  MyTree.Fill()
```

```python
c1 = ROOT.TCanvas( 'c1', 'My First Data example', 200, 10, 700, 900 )
c1.SetFillColor( 18 )
MyTree.Draw("Gauss1>>h1(10,0,10)")
c1.Update()
from ROOT import gROOT 
gROOT.GetListOfCanvases().Draw()
```

```python
MyTree.Draw("Gauss2>>h2(10,0,10)") #primer valor numero de bines
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```python
MyTree.Draw("Plana>>h3(10,0,10)") #primer valor numero de bines
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```python
MyTree.Draw("Gauss1*Gauss2>>h4(10,0,10)")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```python
MyTree.Draw("Gauss1*Plana>>h5(10,0,10)")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```python
MyTree.Draw("Gauss1*Gauss2*Plana>>h6(10,0,10)")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```python
MyTree.Draw("Gauss1>>h7(10,0,10)","4.5<=Gauss1 && Gauss1 <=5.5")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```python
MyTree.Draw("Gauss1:Plana>>h5(10,0,10,10,0,10)","4.5<=Gauss2 && Gauss2 <=5.5","lego")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```python
MyTree.Draw("Gauss1:Gauss2:Plana>>h5(10,0,10,10,0,10,10,0,10)","","lego")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```
